#### **Import Gensim**

```python
from gensim.models import Word2Vec
```

#### **STEP 1: Prepare your data**

```python
# Word2Vec needs a list of sentences
# Each sentence is a list of words (tokens)

sentences = [
    ["the", "cat", "sat", "on", "the", "mat"],
    ["the", "dog", "sat", "on", "the", "floor"],
    ["cats", "and", "dogs", "are", "pets"],
    ["the", "cat", "chased", "the", "mouse"],
    ["the", "dog", "chased", "the", "cat"],
    ["mice", "are", "small", "animals"],
    ["dogs", "and", "cats", "are", "animals"],
    ["the", "mouse", "ran", "away", "from", "the", "cat"],
]
```

#### **STEP 2: Train the model**

```python
model = Word2Vec(
    sentences=sentences,   # your training data
    vector_size=50,        # size of each word vector (embedding dimensions)
    window=3,              # how many words left/right to look at
    min_count=1,           # ignore words that appear less than this
    sg=0,                  # 0 = CBOW,  1 = Skip-Gram
    epochs=100             # how many times to train over the data
)

print("✅ Model trained successfully!\n")
```

#### **STEP 3: Get a word vector**

```python
vector = model.wv["cat"]
print(f"Vector for 'cat' (first 10 values): {vector[:10]}\n")
print(f"Vector size: {len(vector)}\n")
```

#### **STEP 4: Find similar words**

```python
similar = model.wv.most_similar("cat", topn=3)
print("Words most similar to 'cat':")
for word, score in similar:
    print(f"  {word:15s} → similarity score: {score:.4f}")
```

#### **STEP 5: Check similarity between two words**

```python
score = model.wv.similarity("cat", "dog")
print(f"\nSimilarity between 'cat' and 'dog': {score:.4f}")

score2 = model.wv.similarity("cat", "mouse")
print(f"Similarity between 'cat' and 'mouse': {score2:.4f}")
```

#### **STEP 6: Word analogy (king - man + woman = queen concept)**

```python
result = model.wv.most_similar(
    positive=["dog", "cat"],   # words to add
    negative=["mouse"],        # words to subtract
    topn=1
)
print(f"\nAnalogy result (dog + cat - mouse): {result}\n")
```

#### **STEP 7: Save and reload the model**

```python
model.save("word2vec_model.bin")
print("💾 Model saved to word2vec_model.bin")

loaded_model = Word2Vec.load("word2vec_model.bin")
print("📂 Model loaded back successfully!")
print(f"Vocabulary size: {len(loaded_model.wv)}")
print(f"All words: {list(loaded_model.wv.index_to_key)}")
```
