#### **Prompts in langgraph**

```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq


load_dotenv()

model = ChatGroq(
    groq_api_key=os.getenv("GROQ_API_KEY"),
    model="llama-3.1-8b-instant",
    temperature=0.0
)
```

```python
from langchain_core.prompts import ChatPromptTemplate

myprompt = ChatPromptTemplate.from_messages([
    ("system","your a are a helpfull assistant "),
    ("human", "I want to create a post about {topic}"),
])
```

```python
response = myprompt.invoke({"topic" : "the privacy in modern era"})
model.invoke(response).content
```
