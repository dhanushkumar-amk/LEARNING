```python
import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# Step 1 — Import the Gemini model wrapper from LangChain
from langchain_google_genai import ChatGoogleGenerativeAI

# Step 2 — Import the human message type
from langchain_core.messages import HumanMessage

# Step 3 — Create the model object
model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",      # which Gemini model to use
    google_api_key=os.getenv("GOOGLE_API_KEY"), # get API key from .env4
)

# Step 4 — Send a message and get a response
response = model.invoke([
    HumanMessage(content="Explain LangChain in one sentence")
])

# Step 5 — Print the response
print(response.content)
```

```python
# Another way more generic
from langchain.chat_models import init_chat_model


# 3. Initialize the model, explicitly telling it to use the "google_genai" provider
llm_gemini = init_chat_model(
    model="gemini-2.5-flash",
    model_provider="google_genai",
    temperature = 0
)

# 4. Run the code and print the result
response = llm_gemini.invoke("explain about langGraph ?").content
print(response)
```
