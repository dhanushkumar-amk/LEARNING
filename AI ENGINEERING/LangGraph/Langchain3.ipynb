## **Common Code**

```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.prompts import ChatPromptTemplate
from pydantic import BaseModel, Field
from langchain_core.output_parsers import StrOutputParser

load_dotenv()


# model = ChatGroq(
#     groq_api_key=os.getenv("GROQ_API_KEY"),
#     model="llama-3.1-8b-instant",
#     temperature=0.0
# )
```

## **FirstChain**

```python
# task1 (prompt)
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "Your are an helpful assistant"),
    ("human", "{input}")
])
```

```python
# Task 2 (LLM)
model = ChatGroq(
    groq_api_key=os.getenv("GROQ_API_KEY"),
    model="llama-3.1-8b-instant",
    temperature=0.0
)
```

```python
# Task3 String Parser
stringParser = StrOutputParser()
```

## Manual Invocation

```python
template = prompt_template.invoke({"input" : "what is the captial of india"})
result = model.invoke(template)
finalResult = result.content
```

## **Chain Invocation**
----
## **FirstWay**

```python
chain = prompt_template | model | stringParser
chain.invoke({"input" : "what is the captial of india"})
```

## **Another way to create a pipeline**

```python
from langchain_core.runnables import RunnableSequence
chain_1 = RunnableSequence(prompt_template, model, stringParser)
chain_1.invoke({"input" : "what is the captial of india"})
```
