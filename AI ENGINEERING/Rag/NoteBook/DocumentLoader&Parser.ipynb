### **The Document Object**

```python
from langchain_core.documents import Document

# This is what every loader returns
doc = Document(
    page_content="Refunds are processed within 7 business days.",
    metadata={
        "source": "policy.pdf",
        "page": 2
    }
)

print(doc.page_content)  # the text
print(doc.metadata)      # where it came from
```

### **TextLoader — plain .txt files**

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader(
    file_path="hello.txt",
    encoding="utf-8"
)
documents = loader.load()

print(len(documents))             # usually 1 doc per file
print(documents[0].page_content)  # full text
print(documents[0].metadata)      # {'source': 'company_policy.txt'}
```

### **PyPDFLoader — PDF files**

```python
# pip install pypdf
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("Lora.pdf")
documents = loader.load()

# Returns one Document per PAGE
print(len(documents))        # number of pages
print(documents[0].metadata) # {'source': 'annual_report.pdf', 'page': 0}
print(documents[1].metadata) # {'source': 'annual_report.pdf', 'page': 1}
```

```python
from langchain.text_splitter import CharacterTextSplitter

splitter = CharacterTextSplitter(
    separator="\n\n",   # split on paragraph breaks
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_documents(documents)
```
