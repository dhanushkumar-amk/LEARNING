```python
import numpy as np

A = np.array([3, 2])   # "What is the refund policy?"
B = np.array([1, 3])   # "How do I get my money back?"
C = np.array([3, -2])  # "What is the weather today?"
```

### **Cousine Similarity**

```python
def cosine_similarity(v1, v2):
    dot_product = np.dot(v1, v2)
    magnitude = np.linalg.norm(v1) * np.linalg.norm(v2)
    return dot_product / magnitude

print(cosine_similarity(A, B))  # 0.83 → similar ✅
print(cosine_similarity(A, C))  # 0.38 → not similar ❌
```

#### **Dot Product**

```python
def dot_product(v1, v2):
    return np.dot(v1, v2)

print(dot_product(A, B))   # 9  → somewhat similar
print(dot_product(A, C))   # 5  → less similar
```

### **Eculiden distance**

```python
def euclidean_distance(v1, v2):
    return np.linalg.norm(v1 - v2)

print(euclidean_distance(A, B))  # 2.24 → somewhat close
print(euclidean_distance(A, C))  # 4.0  → further apart
```
