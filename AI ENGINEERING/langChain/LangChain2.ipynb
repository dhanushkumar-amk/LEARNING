### Message

```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage

load_dotenv()

model = ChatGroq(
    groq_api_key=os.getenv("GROQ_API_KEY"),
    model="llama-3.1-8b-instant",  # free + fast
    temperature=0.0
)

messages = [
    SystemMessage(content="You are an expert in ai"),
    HumanMessage(content="Explain temperature in AI in simple formate in 2 lines")
]

response = model.invoke(messages)

print(response.content)
```

### Streaming

Streaming (See Response Word by Word)
You know how ChatGPT types word by word? That's streaming.

```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage

load_dotenv()

model = ChatGroq(
    groq_api_key=os.getenv("GROQ_API_KEY"),
    model="llama-3.1-8b-instant",  # free + fast
    temperature=0.0
)

messages = [
    SystemMessage(content="You are an expert in ai"),
    HumanMessage(content="Explain temperature in AI in simple formate in 2 lines")
]

for chunk in model.stream(messages):
    print(chunk.content, end="", flush=True)

print(response.response_metadata)
print(response.id)
```

### Prompt Template

```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.prompts import PromptTemplate   # normal prompt
from langchain_core.prompts import ChatPromptTemplate # add tone also

load_dotenv()


model = ChatGroq(
    groq_api_key=os.getenv("GROQ_API_KEY"),
    model="llama-3.1-8b-instant",
    temperature=0.0
)

userInput = input("Enter the topics : ")
dynamicPrompt = PromptTemplate.from_template("explian the concept of {topic} in 3 lines")
prompt = dynamicPrompt.invoke({"topic" : userInput})

model.invoke(prompt).content
```

```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate

load_dotenv()

model = ChatGroq(
    groq_api_key=os.getenv("GROQ_API_KEY"),
    model="llama-3.1-8b-instant",
    temperature=0.0
)

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "you are a techie assistant"),
    ("user", "give me a topic based on content {topic} in five lines")
])


prompt = prompt_template.invoke({"topic" : "Neural network"})

response = model.invoke(prompt)
print(response.content)
```
