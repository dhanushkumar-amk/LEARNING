## **First Tool**

```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.tools import tool

load_dotenv()

model = ChatGroq(
    groq_api_key=os.getenv("GROQ_API_KEY"),
    model="llama-3.1-8b-instant",
    temperature=0.0
)

# ─────────────────────────────────────
# DEFINE A TOOL
# ─────────────────────────────────────

@tool
def add_numbers(a: int, b: int) -> int:
    """
    Adds two numbers together and returns the result.
    Use this when you need to add or sum two numbers.
    """
    return a + b

result =  add_numbers.invoke({"a" : 10, "b" : 20})
print(result)

# Look at what @tool creates
print(add_numbers.name)         # add_numbers
print(add_numbers.description)  # Adds two numbers together...
print(add_numbers.args)         # {'a': {'type': 'integer'}, 'b': ...}
```

## **Gives tools to Model**

```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage

load_dotenv()

model = ChatGroq(
    groq_api_key=os.getenv("GROQ_API_KEY"),
    model="llama-3.1-8b-instant",
    temperature=0.0
)

@tool
def add(a: int, b: int) -> int:
    """Adds two numbers. Use when you need to add numbers."""
    return a + b

@tool
def get_weather(city: str) -> str:
    """Gets current weather for a city."""
    return "32°C, Sunny"

model_with_tools = model.bind_tools([add, get_weather])

# Ask something that needs a tool
response = model_with_tools.invoke([
    HumanMessage(content="What is 15 + 27?")
])

print("Response type:", type(response))
print("Content:", response.content)
print("Tool calls:", response.tool_calls)
```
