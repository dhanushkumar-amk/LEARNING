```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
a = [10,20,30,40,50]
b = [100,60,70,80,90]

plt.plot(a, b)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("Simple line chart")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

# simple line plot
plt.figure(figsize=(8, 5)) # height and width
plt.plot(x, y, label='sin(x)')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Sins wave")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Scatter plot
np.random.seed(42)

x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.5

plt.figure(figsize=(8, 5))
plt.scatter(x, y, alpha= 1)
plt.xlabel("X values")
plt.ylabel("y values")
plt.title("Scatter plot")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 65, 43, 23]

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='orange', edgecolor='black', linewidth=1.2)
```

```python
# Bar chart horizontal
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 65, 43, 23]

plt.figure(figsize=(8, 6))
plt.barh(categories, values, color='orange', edgecolor='black', linewidth=1.2)
```

```python
# Histrogram

data  = np.random.normal(100, 15, 1000)
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30,color="skyblue", edgecolor="black",)
```
