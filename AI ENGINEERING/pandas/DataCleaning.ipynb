```python
import pandas as pd
```

```python
df = pd.read_json('sample.json')
```

```python
# remove a column
removedColumn  = df.drop(columns=["sepal.length"])
print(removedColumn.head())
```

```python
# remove NaN
dropNaN = df.dropna(subset=["sepal.width"])
print(dropNaN)
```

```python
#  fill the value if a field is NAN
fillNaN = df.fillna({"sepal.width" : 100})
# fillNaN = df.fillna({"sepal.width" : 100, "sepal.length" : 200})
print(fillNaN)
```

```python
# standardization
df['variety'] = df['variety'].str.upper()
print(df)
```

```python
# remove duplicates
df  = df.drop_duplicates()
print(df)
```
