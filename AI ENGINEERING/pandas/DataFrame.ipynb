```python
import pandas as pd
```

```python
# create a dataframe
student = {
    "name" : ["dhanush", "Goms", "shiva", "codeio"],
    "age" : [21,22,54,33],
    "CGpa" : [8.5, 10, 5 ,7.2]
}
df = pd.DataFrame(student)
print(df)
```

```python
# custom Index
df = pd.DataFrame(student, index=["student1", "student2", "student3", "student4"])
print(df)
```

```python
# accesing
print(df.loc["student1"])
print()
print(df.iloc[0])
```

```python
# added a new list to dataframe (column)
df["medium"] = ["english", "tamil", "hindi", "english"]
print(df)
```

```python
# add a new row
student5 = pd.DataFrame({
    'name' : "sanjay",
    "age" : 32,
    "CGpa" : 8.5,
    "medium" : "english"
}, index=["student5"])

df = pd.concat([df, student5])
print(df)
```
