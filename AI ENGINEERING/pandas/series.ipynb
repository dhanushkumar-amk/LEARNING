```python
import pandas as pd
import numpy as np
```

```python
# Create simple series in pandas
#  Series => Caps beacuse it is constructor
age = [22,33,44,55,66]
series = pd.Series(age)
print(series)
```

```python
# Accessing the element by index (location)
print(series.loc[2])
print(series[4])
```

```python
# custom Index
age = [33,4,44,33,22]
customIndex = pd.Series(age, index=["dhanush", "sanjay", "gowtham", "ashwin", "raj"])
print(customIndex)
```

```python
# accessing the element by custom index
print(customIndex['dhanush'])  # direct access
print(customIndex.loc["dhanush"]) # using location
print(customIndex.iloc[0]) # using i location means pass the index value
```

```python
# dictionary to series
dict = {
    "name" : "dhanushkumar",
    "age" : 21,
    "blood" : "B+"
}

dictToSeries = pd.Series(dict)
print(dictToSeries)
```

```python
# Filtering
age = [33,4,44,33,22]
customIndex = pd.Series(age, index=["dhanush", "sanjay", "gowtham", "ashwin", "raj"])
print(customIndex[customIndex > 30])
```

```python
# From a scalar (same value repeated)
s = pd.Series(5, index=['a', 'b', 'c'])
print(s)
```

```python
# From a list
s1 = pd.Series([10, 20, 30])
print("From list:")
print(s1)

# From a list with custom index
s2 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print("\nWith custom index:")
print(s2)

# From a dictionary
s3 = pd.Series(
    {
        'alice': 85,
        'bob': 92,
        'charlie': 78
        }
        )
print("\nFrom dictionary:")
print(s3)

# From a scalar (repeat the same value)
s4 = pd.Series(5, index=['a', 'b', 'c'])
print("\nFrom scalar:")
print(s4)

# From NumPy array
s5 = pd.Series(np.array([1.5, 2.5, 3.5]))
print("\nFrom NumPy array:")
print(s5)
```

```python
s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])

print("Full series:")
print(s)

print("\nAccess by label s['a']:")
print(s['a'])

print("\nAccess by position s.iloc[0]:")
print(s.iloc[0])

print("\nSlice by label s['a':'c']:")
print(s['a':'c'])

print("\nSlice by position s.iloc[0:2]:")
print(s.iloc[0:2])

print("\nMultiple labels s[['a', 'c']]:")
print(s[['a', 'c']])
```

```python
scores = pd.Series([45, 82, 91, 60, 77])

print("Original scores:")
print(scores)

print("\nCondition scores > 70:")
print(scores > 70)

print("\nFiltered scores > 70:")
print(scores[scores > 70])

print("\nMultiple conditions (60 < score < 90):")
print(scores[(scores > 60) & (scores < 90)])

print("\nScores equal to 91:")
print(scores[scores == 91])
```

```python
s = pd.Series([1.0, None, 3.0, None, 5.0])

print("Original series:")
print(s)

print("\nDetect missing values (isnull):")
print(s.isnull())

print("\nDetect non-missing (notnull):")
print(s.notnull())

print("\nCount of missing values:")
print(s.isnull().sum())

print("\nDrop missing values:")
print(s.dropna())

print("\nFill missing with 0:")
print(s.fillna(0))

print("\nFill missing with mean:")
print(s.fillna(s.mean()))

print("\nForward fill (ffill):")
print(s.ffill())

print("\nBackward fill (bfill):")
print(s.bfill())
```

```python
s = pd.Series([10, 20, 30, 40])

print("Original:")
print(s)

print("\nAdd 5 (s + 5):")
print(s + 5)

print("\nMultiply by 2 (s * 2):")
print(s * 2)

print("\nSquare (s ** 2):")
print(s ** 2)

print("\nSquare root (np.sqrt):")
print(np.sqrt(s))

print("\nLog (np.log) — used in loss functions:")
print(np.log(s))

a = pd.Series([1, 2, 3], index=['x', 'y', 'z'])
b = pd.Series([10, 20, 30], index=['x', 'y', 'z'])
print("\nTwo series added (aligns by index):")
print(a + b)
```

```python
names = pd.Series(['Alice Smith', 'bob jones', '  Charlie  ', 'DAVID'])

print("Original:")
print(names)

print("\nLowercase:")
print(names.str.lower())

print("\nUppercase:")
print(names.str.upper())

print("\nStrip whitespace:")
print(names.str.strip())

print("\nReplace space with underscore:")
print(names.str.replace(' ', '_'))

print("\nContains 'bob':")
print(names.str.contains('bob'))

print("\nSplit by space:")
print(names.str.split(' '))

print("\nLength of each string:")
print(names.str.len())

print("\nStarts with 'A':")
print(names.str.startswith('A'))
```

```python
s = pd.Series([10, 20, 30, 40, 50])

print("Original:")
print(s)

print("\nMean:")
print(s.mean())

print("\nMedian:")
print(s.median())

print("\nStandard deviation:")
print(s.std())

print("\nVariance:")
print(s.var())

print("\nMin:")
print(s.min())

print("\nMax:")
print(s.max())

print("\nSum:")
print(s.sum())

print("\nCount (excludes NaN):")
print(s.count())

print("\nDescribe (all stats at once):")
print(s.describe())

print("\nValue counts:")
s2 = pd.Series(['cat', 'dog', 'cat', 'bird', 'dog', 'cat'])
print(s2.value_counts())

print("\nValue counts normalized (percentage):")
print(s2.value_counts(normalize=True))
```

```python
s = pd.Series([30, 10, 50, 20], index=['c', 'a', 'd', 'b'])

print("Original:")
print(s)

print("\nSort by value ascending:")
print(s.sort_values())

print("\nSort by value descending:")
print(s.sort_values(ascending=False))

print("\nSort by index:")
print(s.sort_index())

print("\nTop 2 largest values:")
print(s.nlargest(2))

print("\nTop 2 smallest values:")
print(s.nsmallest(2))
```
